```python
import networkx as nx
import json
import glob
import os
import numpy as np
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from epilepsypcm.utils.make_df import make_df, get_df_list, concat_dfs
from epilepsypcm.utils.outcome_params import engel_score
```

```python
G = nx.Graph()
G.add_node("A")
G.add_nodes_from(['B','C','D','E'])
G.add_edges_from([('A','C'), ('B','D'), ('B','E'), ('C', 'E')])
G.add_edge(*('A','B'))

print(G.nodes()) # returns a list
print(G.edges()) # returns a list

nx.draw(G, with_labels=True)
plt.show() 
```

```python
G = nx.Graph() 
E = [('A', 'B', 2), ('A', 'C', 1), ('B', 'D', 5), ('B', 'E', 3), ('C', 'E', 2)]
G.add_weighted_edges_from(E)
print(G.nodes()) # returns a list
print(G.edges()) # returns a list
```

```python
G = nx.Graph()
G.add_edges_from([('A','B'), ('A','C'), ('B','D'), ('B','E'), ('C', 'E')])

print(G.nodes()) # returns a list
print(G.edges()) # returns a list
```

```python
G = nx.Graph()
G.add_nodes_from([
    (4, {"color": "red"}),
    (5, {"color": "green"}),
])

print(G.nodes()) # returns a list
print(G.edges()) # returns a list
```

```python
DG = nx.DiGraph()
DG.add_edge(2, 1)   # adds the nodes in order 2, 1
DG.add_edge(1, 3)
DG.add_edge(2, 4)
DG.add_edge(1, 2)
print(list(DG.successors(2)))
print(list(DG.edges))

nx.draw(DG, with_labels=True)
plt.show() 
```
